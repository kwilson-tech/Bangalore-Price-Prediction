```python
# Create an instance of Photon

locator = Photon()
```

```python
# Unique location in the dataset

df['Location'].unique()
```

```python
# Length of the unique values

len(df['Location'].unique())
```

```python
# Create a list for the address (add), latitude (lat),longitude(lon), and missing (see what location
# geocode can or can't for the latitude/longitude)
# Missing means 1 if a latitude/longitude was not identified. When missing equals 0, it means a latitude/longitude
# was identified

lat = []
lon = []
location = []
missing = []


for i in range(len(df['Location'].unique())):
    try:
        (address,(latitude,longitude)) = locator.geocode(df['Location'].unique()[i])
        lat.append(latitude)
        lon.append(longitude)
        location.append(df['Location'].unique()[i])
        missing.append(0)
    except:
        lat.append("")
        lon.append("")
        location.append(df['Location'].unique()[i])
        missing.append(1)
```

```python
# Create a DataFrame for the address, latitude, longitude, and missing lists 

identify_ml = pd.DataFrame({'Location':location,'Latitude':lat,'Longitude':lon,'Missing':missing})
```

```python
identify_ml
```

```python
# Identify the locations that don't have a given latitude/longitude and assign it to the 
# variable ml representing missing locations

ml = identify_ml[identify_ml['Missing'] == 1]
```

```python
ml
```

```python
# Remove missing locations 

for i in range(len(ml)):
    df = df[df['Location'] != ml['Location'].values[i]]
```

```python
# Gather known latitude/longitude values  

lat_lon = identify_ml[identify_ml['Missing'] == 0]
```

```python
# Reset index and drop index column

lat_lon = lat_lon.reset_index().drop('index',axis=1)
```

```python
lat_lon
```

```python
# Create columns for Latitude/Longitude in df DataFrame

df['Latitude'] = 0
df['Longitude'] = 0
```

```python
# Reset the index of df and drop the index column

df = df.reset_index().drop('index',axis=1)
```

```python
# Add the given longitude and latitude values to the df dataframe

for i in lat_lon.itertuples(index=False):
    df.at[df.loc[df['Location'] == i[0]].index,'Latitude'] = i[1]
    df.at[df.loc[df['Location'] == i[0]].index,'Longitude'] = i[2]
```
